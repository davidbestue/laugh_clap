```r
library(gdata)
library(lme4)
mydata = read.csv("memory_S.csv")

## lm normal
#lmodel <- lm('recuerdo~tiempo', mydata)
#summary(lmodel)
```

```r
for(i in 1:8) {
    criteria <- c("trick == i")
    jx <-subset(mydata, eval(parse(text = criteria)))
    m <- glmer(memory~duration*time +
        (1 | subject), data = jx, family = binomial, control = glmerControl(optimizer = "bobyqa"),
        nAGQ = 10)

    # print the mod results without correlations among fixed effects
    print(i)
    print(summary(m, corr = FALSE))
}
```
